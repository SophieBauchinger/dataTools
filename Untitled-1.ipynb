Neat lil skewnorm function

```python
def get_skewnorm_pdf(data, **kwargs):
    
    def convert_to_alpha(s):
        """ Converts skewness to alpha parameter - 'a' in scipy.stats """
        d=(np.pi/2*((abs(s)**(2/3))/(abs(s)**(2/3)+((4-np.pi)/2)**(2/3))))**0.5 
        a=((d)/((1-d**2)**.5))
        return(a)
    
    mean = np.mean(data)
    std = np.std(data)
    var = std**2
    skew = stats.skew(data)
    a = convert_to_alpha(skew) 

    # Calculate best fit skewnorm 
    a_f, loc_f, scale_f = stats.skewnorm.fit(data, a, loc = mean, scale = var)

    # Get stats when using new alpha
    mean_ff, var_ff, skew_ff = stats.skewnorm.stats(a_f, loc_f, scale_f, moments='mvs')

    # Get stats when using old alpha (from distribution)
    mean_fi, var_fi, skew_fi = stats.skewnorm.stats(a, loc_f, scale_f, moments='mvs')

    # Display all stats
    import pandas as pd
    df_data = np.array([[mean, mean_fi, mean_ff],
                [std**2, var_fi, var_ff],
                [skew, skew_fi, skew_ff]])
    df = pd.DataFrame(df_data.T, # initial, fi, ff
                    index = ['Initial', 'a_i', 'a_f'], 
                    columns = ['Mean', 'Var', 'Skew'])
    display(df)

    # Plots results
    if kwargs.get('plot', True):
        x = np.linspace(0, 250, int(1e3))
        fig, ax = plt.subplots()
        
        ax.hist(data, bins = kwargs.get('fit_bins', 30),
            edgecolor = 'white', 
            alpha = 0.1, 
            color = 'red',
            density = True)
        
        ax.plot(x, stats.skewnorm.pdf(x, a_f, loc_f, scale_f), label = 'a_fit')
        ax.plot(x, stats.skewnorm.pdf(x, a, loc_f, scale_f), label = 'a_init')
        ax.legend()
        plt.show()

    return mean_fi, var_fi, skew_fi
```

```python
fig, ax = plt.subplots(figsize=(7, 6))

ax_hist = ax.twinx()
scale_hist = 8

boxplot_data = []

for i, bin2d in enumerate([bin_a, bin_b, bin_c]):
    data = bin2d.vstdv.flatten()
    data = data[~np.isnan(data)]
    
    boxplot_data.append(data)

    ax_hist.hist(data, bins = 30,
        edgecolor = 'white', 
        alpha = 0.5,
        density = True,
        bottom = (i+0.3)/scale_hist);
    
    # Jitter the values on the vertical axis
    y = i + np.random.uniform(high=0.2, size=len(data))
    
    ax.scatter(data, y)


# The style of the line that represents the median.
medianprops = {"linewidth": 1, "color": "#a9a9a9", "solid_capstyle": "butt"}
# The style of the box ... This is also used for the whiskers
boxprops = {"linewidth": 1, "color": "#a9a9a9"}

# Positions are shifted now
SHIFT = 0.1
POSITIONS = [0 + SHIFT, 1 + SHIFT, 2 + SHIFT]

ax.set_ylim(-0.1, 3.1)
ymin, ymax = ax.get_ylim()
ax_hist.set_ylim(ymin/scale_hist, ymax/scale_hist)

ax.boxplot(
    boxplot_data, 
    vert=False, 
    positions=POSITIONS, 
    manage_ticks=False,
    showfliers = False, # Do not show the outliers beyond the caps.
    showcaps = False,   # Do not show the caps
    medianprops = medianprops,
    whiskerprops = boxprops,
    boxprops = boxprops
);

# # Finally, add labels and a title
# ax.set_xlabel("Bill length (mm)", fontsize=15)
# ax.set_ylabel("Species", fontsize=15)
# ax.set_title("Bill length by Species", fontsize=18)

# # Change size of tick labels
# ax.tick_params(labelsize=13)
```
