```python
%matplotlib inline
from data import Caribic
import dictionaries as dcts
from plot.binsubs import BinPlotter1D, BinPlotter2D #, BinPlotter3D
```

## Import Data
Import caribic data and detrend the SF6 data with respect to the tropospheric background mixing ratios measured at Mauna Loa. 


```python
caribic = Caribic()
sf6_caribic = dcts.get_subs(col_name='SF6')
o3_caribic = dcts.get_subs(col_name='int_O3')

caribic.detrend_substance(sf6_caribic)
```

## Calculate Plotters from Caribic data

```python
bp1 = BinPlotter1D(caribic)
bp2 = BinPlotter2D(caribic)
```

## Plotting

### Seasonal 2D Binned plots

#### $\text{O}_3$ vmean
1) Latitude vs. $\Theta$
2) Eqlat vs. $\Theta$
3) Latitude vs. $\Delta\Theta_{therm}$
4) Eqlat vs. $\Delta\Theta_{therm}$

```python
# 1. Lat vs. Theta
bp2.seasonal_2d_plots(dcts.get_subs(col_name='int_O3'),
dcts.get_coord(col_name='geometry.y'),
dcts.get_coord(col_name='int_ERA5_THETA'),
'vmean', xbsize=5, ybsize=5)
```

```python
# 2. Eqlat vs. Theta
bp2.seasonal_2d_plots(dcts.get_subs(col_name='int_O3'),
dcts.get_coord(col_name='int_ERA5_EQLAT'),
dcts.get_coord(col_name='int_ERA5_THETA'),
'vmean', xbsize=5, ybsize=5)
```

```python
# 3. Lat vs. Theta wrt thermal TP 
bp2.seasonal_2d_plots(dcts.get_subs(col_name='int_O3'),
dcts.get_coord(col_name='geometry.y'),
caribic.tp_coords(vcoord='pt', rel_to_tp=True, model='ERA5', tp_def='therm')[0],
'vmean', xbsize=5, ybsize=5)
```

```python
# 4. Eqlat vs. Theta wrt thermal TP
bp2.seasonal_2d_plots(dcts.get_subs(col_name='int_O3'),
dcts.get_coord(col_name='int_ERA5_EQLAT'),
caribic.tp_coords(vcoord='pt', rel_to_tp=True, model='ERA5', tp_def='therm')[0],
'vmean', xbsize=5, ybsize=5)
```
